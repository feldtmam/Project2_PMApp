```python
# import dependencies 
import sqlite3 as sql
import pandas as pd
from datetime import date
```

```python
# create connection to database 
database = ('pmappdb.db')
connection = sql.connect(database)
```

```python
query = ("SELECT * FROM task")
df = pd.read_sql_query(query, connection)
df.head()
```

```python
df['start_date'] = pd.to_datetime(df['start_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
df.head()
```

```python
df.set_index('task_id', inplace=True)
df.head()
```

```python
proj_demo = df['project_title']=="Project Demo"
demo_df = df[proj_demo]
demo_df.head()
```

```python
demo_progress = demo_df[["resource_id", "status"]].groupby('status').count()
demo_progress = demo_progress.rename(columns={'resource_id': 'total_tasks'})
demo_progress
```

```python
demo_progress.reset_index().to_json(r'demo_progress.json', orient = 'records')
```

```python
demo_resource_task_total = demo_df[["resource_id", "status"]].groupby('resource_id').count()
demo_resource_task_total = demo_resource_task_total.rename(columns={'status': 'total_tasks'})
demo_resource_task_total
```

```python
demo_resource_task_total.reset_index().to_json(r'demo_task_total.json', orient = 'records')
```

```python
# math_12th = merged_data.loc[merged_data["grade"] == "12th"].groupby("school_name")["math_score"].mean()
inprogress = demo_df.loc[demo_df["status"] == "In Progress"].groupby("resource_id")["status"].count()
completed = demo_df.loc[demo_df["status"] == "Completed"].groupby("resource_id")["status"].count()
deferred = demo_df.loc[demo_df["status"] == "Deferred"].groupby("resource_id")["status"].count()
notstarted = demo_df.loc[demo_df["status"] == "Not Started"].groupby("resource_id")["status"].count()
```

```python
demo_status_df = pd.DataFrame ({"In Progress": inprogress,
                           "Completed": completed,
                           "Deferred": deferred,
                            "Not Started": notstarted})
demo_status_df
```

```python
demo_status_df.reset_index().to_json(r'demo_status_totals.json', orient = 'records')
```
